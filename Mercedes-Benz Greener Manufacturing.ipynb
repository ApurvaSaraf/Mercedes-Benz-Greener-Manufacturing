```python
# Perform dimensionality reduction.
from sklearn.decomposition import PCA
   
PCa = PCA (n_components = 2)  
   
X_train_f = PCa.fit_transform(final_train)  
X_test_f = PCa.fit_transform(final_test)
y = df_train['y']
```

# Predict your test_df values using XGBoost.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_f, y, test_size=0.33, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_squared_error
XGBr=XGBRegressor(random_state=42)

model=XGBr.fit(X_train,y_train)
y_prd=model.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_prd))
print(r2_score(y_test,y_prd))
```

```python
pred=pd.DataFrame({'y_test':y_test,'y_prd':y_prd})
pred
```
